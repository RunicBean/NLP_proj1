```python
import numpy as np
import pandas as pd
```

```python
def smooth_kernel_3(sentenceArray):
    out=[]
    cnnArray=np.array([[1/3],[1/3],[1/3]])
    for i in range (2,len(sentenceArray)):
        c=np.dot(sentenceArray[i-2:i+1],cnnArray)
        if len(out)==0 or len(out)==len(sentenceArray)-2:
            out=out+list(c)
      
        out=out+list(c)
    for i in range (0,len(out)):
        out[i]=round(out[i],3)
        
    #out.append((sentenceArray[-1]+sentenceArray[-2])/2)
    return out
```

```python
def smooth_kernel_4(sentenceArray):
    out=[]
    cnnArray=np.array([1/4,1/4,1/4,1/4])
    for i in range (1,len(sentenceArray)-2):
        c=np.dot(sentenceArray[i-1:i+3],cnnArray)
        #print(sentenceArray[i-1:i+3],c)
        if len(out)==0:
            out=out+[round(c,3)]
        if len(out)==len(out)==len(sentenceArray)-3:
            out=out+[round(c,3)]+[round(c,3)]
        out=out+[round(c,3)]
        #print(out)
    #out.append((sentenceArray[-1]+sentenceArray[-2])/2)
    return out

#a=np.array([0.1,0.5,0.3,0.7,0.8,0.9,0.5,0.7,0.4])
##b=np.array([[1/4],[1/4],[1/4],[1/4]])
#tmp=smooth_kernel_4(a)
#tmp
```

```python
def smooth_kernel_5(sentenceArray):
    out=[]
    cnnArray=np.array([1/5,1/5,1/5,1/5,1/5])
    for i in range (2,len(sentenceArray)-2):
        c=np.dot(sentenceArray[i-2:i+3],cnnArray)
        #print(sentenceArray[i-1:i+3],c)
        if len(out)==0:
            out=out+[round(c,3)]+[round(c,3)]
        if len(out)==len(out)==len(sentenceArray)-3:
            out=out+[round(c,3)]+[round(c,3)]
        out=out+[round(c,3)]
        #print(out)
    #out.append((sentenceArray[-1]+sentenceArray[-2])/2)
    return out
#a=np.array([0.1,0.5,0.3,0.7,0.8,0.9,0.5,0.7,0.4])
#b=np.array([[1/4],[1/4],[1/4],[1/4]])
#tmp=smooth_kernel_5(a)
#tmp
```

```python
def get_concise(input_file,kernel):
    processed = pd.read_csv(input_file)
    
    processed["new_score_3"]=smooth_kernel_3(processed["score"])
    processed["new_score_4"]=smooth_kernel_4(processed["score"])
    processed["new_score_5"]=smooth_kernel_5(processed["score"])
    
    processed.columns=['idx','score','content','kernel3','kernel4','kernel5']
    processed.sort_values(kernel,inplace=True)
    #sort by kernel score

    top_ranked=int(len(processed)/4) ##select the sentenses with top 1/4 importance
    tmp=processed[0:top_ranked]
    tmp.sort_values("idx",inplace=True)
    tmp
    return tmp["content"]
        
list1=get_concise("doc_sim.csv","kernel3")
#the first input is input csv file
#the second input can be selected from:
#kernel4
#kernel5
#kernel3
```
